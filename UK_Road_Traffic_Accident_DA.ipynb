### **1.2. Dữ liệu vào (Input Data)**
+ **A. Tệp Accident_Information (Thông tin tai nạn):**
  + **Nhóm 1: Thông tin thời gian:**
    + Date: Ngày xảy ra tai nạn.
    + Time: Thời gian xảy ra tai nạn.

    + Year: Năm xảy ra tai nạn.
  + **Nhóm 2: Thông tin địa lí và vị trí:**

  + **Nhóm 3: Đặc điểm đường và giao lộ:**
    + 1st_Road_Class, 1st_Road_Number: Hạng và số hiệu đường chính
    + 2nd_Road_Class, 2nd_Road_Number: Hạng và số hiệu đường thứ hai
    + Road_Type: Loại đường
    + Speed_Limit: Tốc độ giới hạn
    + Junction_Control, Junction_Detail: Cách kiểm soát và chi tiết giao lộ
  + **Nhóm 4: Điều kiện môi trường và hiện trường:**
    + Light_Conditions: Điều kiện ánh sáng
    + Weather_Conditions: Điều kiện thời tiết
    + Road_Surface_Conditions: Tình trạng bề mặt đường
    + Special_Conditions_at_Site: Điều kiện đặc biệt tại hiện trường
    + Carriageway_Hazards: Nguy cơ/chướng ngại vật trên đường xe chạy
  + **Nhóm 5: Hậu quả và đáp ứng:**
    + Number_of_Vehicles: Số phương tiện liên quan
    + Number_of_Casualties: Số lượng nạn nhân
    + Did_Police_Officer_Attend_Scene_of_Accident: Cảnh sát có mặt tại hiện trường không?
    + Police_Force: Vùng Địa lý mà Lực lượng Cảnh sát đó phụ trách quản lý
    + Pedestrian_Crossing-Human_Control: Người điều khiển vạch sang đường
    + Pedestrian_Crossing-Physical_Facilities: Cơ sở vật chất cho người đi bộ
    + InScotland: Tai nạn có xảy ra ở Scotland không

+ **B. Tệp Vehicle_Information (Thông tin phương tiện)**

  + **Nhóm 1: Thông tin phương tiện:**
    + Vehicle_Type: Loại xe
    + make, model: Hãng xe và mẫu xe
    + Age_of_Vehicle: Tuổi của xe
    + Engine_Capacity_.CC.: Dung tích động cơ
    + Propulsion_Code: Loại động cơ
    + Was_Vehicle_Left_Hand_Drive: Có phải xe tay lái nghịch không
    + Year: Năm sản xuất phương tiện
  + **Nhóm 2: Thông tin người lái và hành trình:**
    + Sex_of_Driver: Giới tính của tài xế
    + Age_Band_of_Driver: Nhóm tuổi của tài xế
    + Journey_Purpose_of_Driver: Mục đích chuyến đi
    + Driver_Home_Area_Type: Loại khu vực cư trú của tài xế
    + Driver_IMD_Decile: Mức độ thiếu thốn của tài xế
  + **Nhóm 3: Động lực học va chạm:**
    + Vehicle_Manoeuvre: Thao tác của xe trước khi va chạm
    + X1st_Point_of_Impact: Điểm va chạm đầu tiên
    + Vehicle_Leaving_Carriageway: Xe có rời khỏi phần đường xe chạy không
    + Skidding_and_Overturning: Có trượt bánh và lật xe không
    + Hit_Object_in_Carriageway: Vật thể bị va chạm trên làn đường
    + Hit_Object_off_Carriageway: Vật thể bị va chạm ngoài làn đường
    + Junction_Location: Vị trí tại giao lộ
    + Towing_and_Articulation: Tình trạng kéo theo hoặc đầu kéo
    + Vehicle_Location.Restricted_Lane: Vị trí xe trên làn đường bị hạn chế

###**Bước 3: Hợp nhất dữ liệu**

**- Do mỗi vụ tai nạn có thể liên quan đến nhiều phương tiện khác nhau, bảng Vehicle cần được tổng hợp theo khóa Accident_Index để đảm bảo mỗi bản ghi trong dataset cuối cùng đại diện cho một vụ tai nạn duy nhất, tránh hiện tượng nhân bản bản ghi khi hợp nhất dữ liệu, đồng thời vẫn giữ được thông tin đặc trưng của các phương tiện liên quan đến mỗi vụ tai nạn. Quá trình tổng hợp, sẽ áp dụng theo quy ước:**

  - Đối với các biến phân loại (categorical):
Giá trị đại diện được xác định bằng mode (giá trị xuất hiện nhiều nhất), nhằm phản ánh đặc điểm phổ biến nhất của các phương tiện liên quan trong cùng một vụ tai nạn.
  - Đối với các biến dạng số (numerical):
Sử dụng median (thay vì mean) vì nó không bị ảnh hưởng bởi các giá trị ngoại lai, đồng thời phản ánh xu hướng trung tâm ổn định hơn của dữ liệu.

```python
import polars as pl

# Đổi tên cột để sử dụng
vehicle_clean = vehicle_clean.rename(columns={'Engine_Capacity_.CC.':'Engine_Capacity_CC'})
df = pl.from_pandas(vehicle_clean)

# Aggregate theo Accident_Index
vehicle_agg_pl = df.group_by("Accident_Index").agg([
    pl.col("Vehicle_Type").mode().first().alias("Vehicle_Type"),
    pl.col("Vehicle_Manoeuvre").mode().first().alias("Vehicle_Manoeuvre"),
    pl.col("Vehicle_Leaving_Carriageway").mode().first().alias("Vehicle_Leaving_Carriageway"),
    pl.col("Age_Band_of_Driver").mode().first().alias("Age_Band_of_Driver"),
    pl.col("Sex_of_Driver").mode().first().alias("Sex_of_Driver"),
    pl.col("Age_of_Vehicle").median().alias("Age_of_Vehicle"),
    pl.col("Engine_Capacity_CC").median().alias("Engine_Capacity_CC")
])

vehicle_agg = vehicle_agg_pl.to_pandas()
print(f"Hoàn thành! Kết quả có {len(vehicle_agg)} dòng")
```

**- Sau khi đã tổng hợp dữ liệu phương tiện theo Accident_Index, chúng ta có bảng vehicle_agg với mỗi dòng đại diện cho một vụ tai nạn duy nhất đồng thời chứa thông tin tổng hợp của các phương tiện liên quan.**

**- Bước tiếp theo, chúng ta sẽ gộp bảng Accident với bảng Vehicle đã tổng hợp để tạo dataset cuối cùng phục vụ cho phân tích khám phá dữ liệu (EDA) và xây dựng mô hình. Với Khóa chính để merge là Accident_Index, sử dụng phương pháp LEFT JOIN để đảm bảo giữ lại tất cả các vụ tai nạn trong accidents_clean kể cả những vụ không có thông tin phương tiện tương ứng.**

```python
merged_data = pd.merge(
    accidents_clean, # bảng tai nạn
    vehicle_agg,  # bảng vehicle đã aggregate
    on='Accident_Index',
    how='left'
)

merged_data.info()
display(merged_data.head(5))
```

###**Bước 4: Xử lí dữ liệu thiếu (Missing values)**

**- Kiểm tra các giá trị thiếu (Missing Values) của biến số:**

```python
# Lọc các biến số
numerical_cols = merged_data.select_dtypes(include=['float64','int64']).columns

# Thống kê missing chỉ cho biến số
missing_stats = merged_data[numerical_cols].isnull().sum()
missing_percent = (missing_stats / len(merged_data)) * 100

missing_df = pd.DataFrame({
    'Số lượng missing': missing_stats,
    'Tỷ lệ %': missing_percent
}).sort_values('Số lượng missing', ascending=False)

# Chỉ hiển thị những cột có missing
missing_df = missing_df[missing_df['Số lượng missing'] > 0]

display(missing_df)
```

**- Kiểm tra các giá trị thiếu (Missing Values) của biến phân loại:**

```python
# Xác định các biến phân loại
categorical_cols = merged_data.select_dtypes(include=['object']).columns

# Kiểm tra missing values cho biến phân loại
categorical_missing = merged_data[categorical_cols].isnull().sum()
categorical_missing_percent = (categorical_missing / len(merged_data)) * 100
categorical_missing_df = pd.DataFrame({
    'Số lượng missing': categorical_missing,
    'Tỷ lệ %': categorical_missing_percent
}).sort_values('Số lượng missing', ascending=False)
categorical_missing_df = categorical_missing_df[categorical_missing_df['Số lượng missing'] > 0]
display(categorical_missing_df)
```

**Thực hiện xử lí các giá trị thiếu (Missing values) cho cả biến số và biến phân loại theo quy ước:**

- Tỉ lệ missing >50%: xóa cột

- Tỉ lệ mising 5-50%: impute (median) nếu là biến số, impute nhãn (Missing) nếu là biến phân loại.

- Tỉ lệ mising <5%: xóa dòng.

```python
# Xử lý biến số
for col in numerical_cols:
    missing_rate = merged_data[col].isnull().mean()

    if missing_rate == 0:
        continue  # Bỏ qua cột không missing

    if missing_rate > 0.5:
        merged_data = merged_data.drop(columns=[col])
        print(f"[DROP COL] {col}: missing rate = {missing_rate:.5%} >50%, drop column")

    elif 0.05 <= missing_rate <= 0.5:
        median_val = merged_data[col].median()
        merged_data[col] = merged_data[col].fillna(median_val)
        print(f"[IMPUTE] {col}: missing rate = {missing_rate:.5%}, fillna with median = {median_val}")

    else:  # missing_rate <5%
        merged_data = merged_data[merged_data[col].notnull()]
        print(f"[DROP ROW] {col}: missing rate = {missing_rate:.5%} <5%, drop rows with NaN")

# Xử lý biến phân loại
for col in categorical_cols:
    missing_rate = merged_data[col].isnull().mean()

    if missing_rate == 0:
        continue  # Bỏ qua cột không missing

    if missing_rate > 0.5:
        merged_data = merged_data.drop(columns=[col])
        print(f"[DROP COL] {col}: missing rate = {missing_rate:.5%} >50%, drop column")

    elif 0.05 <= missing_rate <= 0.5:
        merged_data[col] = merged_data[col].fillna('Missing')
        print(f"[IMPUTE] {col}: missing rate = {missing_rate:.5%}, fillna with 'Missing'")

    else:  # missing_rate <5%
        merged_data = merged_data[merged_data[col].notnull()]
        print(f"[DROP ROW] {col}: missing rate = {missing_rate:.5%} <5%, drop rows with NaN")
```

**- Tiếp theo, thực hiện kiểm tra các giá trị không rõ ràng của biến phân loại:**

```python
categorical_cols = merged_data.select_dtypes(include=['object']).columns
total_records = len(merged_data)

special_values = ["Data missing or out of range", "Unknown", "Unallocated", "Other", "Other/Not known", "Not known"]

special_found = False

for col in categorical_cols:
    col_series = merged_data[col].astype(str).str.strip()
    col_series_lower = col_series.str.lower()

    # Đếm từng giá trị đặc biệt
    counts = {}
    for val in special_values:
        if val.lower() in ["unknown", "unallocated"]:
            counts[val] = (col_series_lower == val.lower()).sum()
        else:
            counts[val] = (col_series == val).sum()

    if sum(counts.values()) == 0:
        continue

    if not special_found:
        print(f"{'CỘT':<35} {'GIÁ TRỊ':<40} {'SỐ LƯỢNG':>10}   {'TỶ LỆ':>7}")
        print("-"*100)
        special_found = True

    # In từng giá trị
    for val, count in counts.items():
        if count > 0:
            print(f"{col:<35} {val:<40} {count:>10,}   {count/total_records*100:>6.4f}%")

    print("-"*100)
```

**- Thực hiện gom các giá trị này thành một category chung (Missing). Bao gồm các giá trị: Data missing or out of range, Unknown, Other, Other/Not known, Not known. Sau khi gom, xử lý theo tỉ lệ missing:**

- Nếu tỉ lệ "Missing" >5%: [KEEP] giữ nguyên category "Missing".
- Nếu tỉ lệ "Missing" <5%: [IMPUTE] mode - giá trị xuất hiện nhiều nhất.

```python
missing_labels = [
    "Data missing or out of range",
    "Unknown",
    "Not known",
    "Other",
    "Other/Not known (2005-10)",
    "Unallocated"
]

categorical_cols = merged_data.select_dtypes(include=['object']).columns

for col in categorical_cols:
    # Chuẩn hóa text
    merged_data[col] = merged_data[col].astype(str).str.strip()

    # Gom các nhãn đặc biệt thành 'Missing'
    merged_data[col] = merged_data[col].replace(
        {label: "Missing" for label in missing_labels}
    )

    # Tính tỉ lệ Missing
    missing_rate = (merged_data[col] == "Missing").mean()

    if missing_rate > 0.05:
        # Giữ nguyên category Missing
        print(f"[KEEP] {col}: {missing_rate:.2%} → kept as separate 'Missing' category")
    elif missing_rate > 0:
        # Impute bằng mode
        mode_val = merged_data[col].mode()[0]
        merged_data[col] = merged_data[col].replace("Missing", mode_val)
        print(f"[IMPUTE] {col}: {missing_rate:.2%} → replaced Missing with mode = {mode_val}")
    # Nếu missing_rate == 0, không làm gì
```

**Thực hiện xử lí các giá trị Outliers bất thường của 2 biến **Age_of_Vehicle** và **Engine_Capacity_.CC. theo quy ước sử dụng IQR Capping:****
+ lower bound = Q1 - 1.5 * IQR
+ upper bound = Q3 + 1.5 * IQR
+ Giá trị < lower bound → gán bằng lower bound
+ Giá trị > upper bound → gán bằng upper bound

```python
num_cols_cap = ['Age_of_Vehicle', 'Engine_Capacity_CC']

summary_list = []

for col in num_cols_cap:
    Q1 = merged_data[col].quantile(0.25)
    Q3 = merged_data[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    print(f"Column: {col}")
    print(f"Lower bound: {lower_bound:.2f}, Upper bound: {upper_bound:.2f}")

    # Capping
    merged_data[col] = np.where(
        merged_data[col] < lower_bound, lower_bound,
        np.where(merged_data[col] > upper_bound, upper_bound, merged_data[col])
    )

    # Lưu summary sau capping để hiển thị trong bảng
    desc = merged_data[col].describe()
    summary_list.append({
        "Variable": col,
        "Count": desc["count"],
        "Mean": desc["mean"],
        "Std": desc["std"],
        "Min": desc["min"],
        "25%": desc["25%"],
        "50%": desc["50%"],
        "75%": desc["75%"],
        "Max": desc["max"]
    })

summary_df = pd.DataFrame(summary_list)
display(summary_df)
```

### **Bước 7: Feature Engineering cơ bản**

**Mục tiêu:** nhằm tạo ra các biến mới có ý nghĩa từ các cột thời gian để hỗ trợ phân tích. Việc tạo các biến này không chỉ làm giàu thông tin cho dữ liệu mà còn hỗ trợ trực quan hóa và mô hình hóa sau này, đồng thời giúp phát hiện các xu hướng và đặc điểm ẩn trong dữ liệu tai nạn.

```python
# Tạo biến Time_period theo Hour
def assign_time_period(hour):
    if pd.isna(hour):
        return None
    if 5 <= hour < 10:
        return 'Sáng sớm'
    elif 10 <= hour < 15:
        return 'Ban ngày'
    elif 15 <= hour < 19:
        return 'Chiều tối'
    elif 19 <= hour < 23:
        return 'Tối'
    else:
        return 'Khuya'

merged_data['Time_period'] = merged_data['Hour'].apply(assign_time_period)
```

**Mô tả:** Dựa trên cột **Hour** trích xuất từ cột **Time**, ta phân loại giờ trong ngày thành các khung thời gian có ý nghĩa như “Sáng sớm”, “Ban ngày”, “Chiều tối”, “Tối” và “Khuya”, giúp phân tích tai nạn theo thời điểm trong ngày.

+ Kết quả này cho thấy các va chạm nhẹ không xảy ra ngẫu nhiên theo thời gian, mà có xu hướng tập trung rõ rệt vào các ngày làm việc và đặc biệt là khung giờ cao điểm **Chiều tối**. Đây là dấu hiệu đặc trưng của những tai nạn xuất phát từ mật độ giao thông cao và hành vi vội vã.

Khi thời điểm rủi ro đã được xác định, phân tích tiếp theo chuyển sang góc độ **KHÔNG GIAN (WHERE)** nhằm trả lời câu hỏi: ***Các va chạm nhẹ chủ yếu xảy ra ở đâu – trong khu vực đô thị đông đúc hay trên các tuyến đường nông thôn thưa thớt?***

```python
# Lọc va chạm nhẹ
slight_data = merged_data[
    merged_data['Accident_Severity'] == 'Slight'
]

plt.figure(figsize=(7, 5))

ax = sns.countplot(
    data=slight_data,
    x='Urban_or_Rural_Area',
    color='#D6C1A3'
)

plt.title("Phân bố va chạm nhẹ theo khu vực", fontsize=14)
plt.xlabel("Khu vực", fontsize=11)
plt.ylabel("Số vụ va chạm nhẹ", fontsize=11)

# Ghi số lượng + %
total = len(slight_data)
max_count = max(p.get_height() for p in ax.patches)

for p in ax.patches:
    count = int(p.get_height())
    pct = count / total * 100

    # Highlight khu vực chiếm ưu thế
    if count == max_count:
        p.set_color('#6B4F3F')

    ax.text(
        p.get_x() + p.get_width() / 2,
        count + total * 0.005,
        f"{count:,} ({pct:.2f}%)",
        ha='center',
        va='bottom',
        fontsize=10
    )

plt.tight_layout()
plt.show()
```

**→ Nhận xét:** Dữ liệu thể hiện một sự phân bố không gian rất rõ: phần lớn các va chạm nhẹ tập trung ở khu vực **THÀNH THỊ (Urban)** với 67.04%, tương đương gần nửa triệu vụ. Tỷ lệ này cao gấp đôi so với khu vực **NÔNG THÔN (Rural).** Điều này củng cố mạnh mẽ mối liên hệ giữa mật độ giao thông và tai nạn nhẹ. Rủi ro không nằm ở những con đường vắng, mà chính ở những không gian đô thị đông đúc, nơi xe cộ tiếp xúc gần và liên tục.

---

Khi các va chạm nhẹ tập trung áp đảo trong khu vực đô thị, câu hỏi tiếp theo không còn là ở đâu, mà là ở loại hạ tầng nào: ***Trong môi trường đô thị, loại đường nào đang ghi nhận nhiều va chạm nhẹ nhất?***

```python
# Lọc va chạm nhẹ trong khu vực đô thị
urban_slight = merged_data[
    (merged_data['Accident_Severity'] == 'Slight') &
    (merged_data['Urban_or_Rural_Area'] == 'Urban')
]

plt.figure(figsize=(10, 5))

ax = sns.countplot(
    data=urban_slight,
    x='Road_Type',
    order=urban_slight['Road_Type'].value_counts().index,
    color='#BCAAA4'
)

plt.title("Phân bố va chạm nhẹ theo loại đường (Khu vực đô thị)", fontsize=14)
plt.xlabel("Loại đường", fontsize=11)
plt.ylabel("Số vụ va chạm nhẹ", fontsize=11)

# Ghi số lượng + %
total = len(urban_slight)
max_count = max(p.get_height() for p in ax.patches)

for p in ax.patches:
    count = int(p.get_height())
    pct = count / total * 100

    # Highlight loại đường chiếm ưu thế
    if count == max_count:
        p.set_color('#6B4F3F')

    ax.text(
        p.get_x() + p.get_width() / 2,
        count + total * 0.004,
        f"{count:,} ({pct:.2f}%)",
        ha='center',
        va='bottom',
        fontsize=9
    )

plt.tight_layout()
plt.show()
```

**→ Nhận xét:** Kết quả cho thấy một thực tế nổi bật **ĐƯỜNG ĐƠN (Single carriageway)** là loại hạ tầng ghi nhận nhiều va chạm nhẹ nhất với tỷ lệ áp đảo gần 78% cao hơn gấp nhiều lần so với bất kỳ loại đường nào khác trong đô thị. Điều này phản ánh các va chạm nhẹ chủ yếu diễn ra trên những tuyến đường hai chiều, nhiều điểm rẽ và có không gian lưu thông hạn chế.

---

Từ bối cảnh hạ tầng trên, một câu hỏi tiếp theo được đặt ra: ***Trên các tuyến đường đơn trong đô thị, các va chạm nhẹ có tập trung nhiều hơn tại các giao lộ, vòng xuyến hay trên những đoạn đường không có nút giao?***

```python
# Lọc va chạm nhẹ – đô thị – đường đơn
junction_data = merged_data[
    (merged_data['Accident_Severity'] == 'Slight') &
    (merged_data['Urban_or_Rural_Area'] == 'Urban') &
    (merged_data['Road_Type'] == 'Single carriageway')
]

# Lấy TOP 5 loại giao lộ
top_junctions = (
    junction_data['Junction_Detail']
    .value_counts()
    .head(5)
)

top_junction_df = top_junctions.reset_index()
top_junction_df.columns = ['Junction_Detail', 'Accident_Count']

total = top_junction_df['Accident_Count'].sum()
max_count = top_junction_df['Accident_Count'].max()

plt.figure(figsize=(10, 5))

ax = sns.barplot(
    data=top_junction_df,
    y='Junction_Detail',
    x='Accident_Count',
    color='#D6C1A3'
)

# Highlight điểm nóng nhất
for i, row in top_junction_df.iterrows():
    if row['Accident_Count'] == max_count:
        ax.patches[i].set_color('#6B4F3F')

# Ghi số lượng + %
for i, row in top_junction_df.iterrows():
    pct = row['Accident_Count'] / total * 100
    ax.text(
        row['Accident_Count'] + total * 0.01,
        i,
        f"{row['Accident_Count']:,} ({pct:.2f}%)",
        va='center',
        fontsize=9
    )

plt.title(
    "Top 5 loại hình giao lộ ghi nhận nhiều va chạm nhẹ nhất\n(Đô thị – Đường đơn)",
    fontsize=14
)
plt.xlabel("Số vụ va chạm nhẹ", fontsize=11)
plt.ylabel("Loại hình giao lộ", fontsize=11)

plt.tight_layout()
plt.show()
```

**→ Nhận xét:** Dữ liệu cho thấy một nghịch lý đáng chú ý: trong khi gần 1/2 va chạm nhẹ xảy ra tại các **NGÃ BA, NGÃ TƯ (T or Staggered junction)** – nơi luồng giao thông giao cắt phức tạp – thì vẫn có tới gần 1/3 số vụ diễn ra trên các đoạn đường **KHÔNG CÓ GIAO LỘ (Not at junction or within 20 metres)**. Điều này cho thấy rủi ro va chạm không chỉ tập trung ở những điểm giao cắt đông đúc, mà còn hiện diện ngay cả trên những đoạn đường tưởng chừng an toàn của các tuyến đường đơn trong đô thị.

---

Khi thời điểm và không gian của các va chạm nhẹ đã được xác định rõ, câu hỏi cuối cùng và quan trọng nhất được đặt ra không còn là khi nào hay ở đâu, mà là **ai (WHO)** đang tham gia vào những sự cố này và họ đang điều khiển **loại phương tiện nào (WHAT)**, trước hết chúng ta sẽ tập trung vào nhóm tuổi tài xế: ***Nhóm tuổi nào thường xuyên tham gia vào các va chạm nhẹ, và liệu có nhóm tuổi nào chiếm áp đảo so với các nhóm còn lại?***

```python
# Lọc va chạm nhẹ
slight_data = merged_data[merged_data['Accident_Severity'] == 'Slight']
age_valid = slight_data[~slight_data['Age_Band_of_Driver'].isin(['Missing'])]

# Sắp xếp theo tần suất giảm dần
age_order = age_valid['Age_Band_of_Driver'].value_counts().index

plt.figure(figsize=(10,6))

ax = sns.countplot(
    data=age_valid,
    y='Age_Band_of_Driver',
    order=age_order,
    color='#BDBDBD'
)

plt.title("Phân bố va chạm nhẹ theo nhóm tuổi tài xế", fontsize=14)
plt.xlabel("Số vụ va chạm nhẹ")
plt.ylabel("Nhóm tuổi tài xế")

total = len(age_valid)
max_count = max(p.get_width() for p in ax.patches)

for p in ax.patches:
    count = int(p.get_width())
    pct = count / total * 100

    # Highlight nhóm chiếm nhiều nhất
    if count == max_count:
        p.set_color('#616161')

    ax.text(
        count + total*0.004,
        p.get_y() + p.get_height()/2,
        f"{count:,} ({pct:.2f}%)",
        va='center',
        fontsize=9
    )

plt.tight_layout()
plt.show()
```

**→ Nhận xét:** Nhóm tài xế từ 26 đến 45 tuổi (bao gồm cả hai nhóm 26-35 và 36-45) cùng chiếm gần 43% tổng số va chạm nhẹ, trở thành nhóm có tỷ lệ cao áp đảo. Điều này cho thấy độ tuổi trung niên đang trong giai đoạn hoạt động giao thông cao độ – thường là lực lượng lao động chính, tham gia giao thông liên tục vào giờ cao điểm các ngày làm việc – chính là đối tượng trung tâm của các vụ va chạm nhỏ, chứ không phải nhóm thanh thiếu niên thiếu kinh nghiệm (16-25 tuổi, chiếm ~26%) hay nhóm lớn tuổi (trên 56 tuổi, chiếm ~22%).

---

Sau khi đã xác định được nhóm tuổi chính từ 26 đến 45 là đối tượng trung tâm, một câu hỏi tiếp theo được đặt ra: ***Trong chính nhóm tuổi trung niên này, giới tính nào đang chiếm đa số trong các vụ va chạm nhẹ?***

```python
# Lọc va chạm nhẹ
slight_data = merged_data[merged_data['Accident_Severity'] == 'Slight']

# Lọc nhóm tuổi trung niên
mid_age_data = slight_data[slight_data['Age_Band_of_Driver'].isin(['26 - 35', '36 - 45'])]
mid_age_data = mid_age_data[mid_age_data['Sex_of_Driver'] != 'Missing']

plt.figure(figsize=(6,4))

ax = sns.countplot(
    data=mid_age_data,
    x='Sex_of_Driver',
    order=mid_age_data['Sex_of_Driver'].value_counts().index,
    palette=['#6baed6', '#fb8072']
)

plt.title("Phân bố giới tính tài xế trong nhóm tuổi trung niên\n (Va chạm nhẹ)", fontsize=14)
plt.xlabel("Giới tính tài xế")
plt.ylabel("Số vụ va chạm nhẹ")

# Ghi số lượng + %
total = len(mid_age_data)

for p in ax.patches:
    count = int(p.get_height())
    pct = count / total * 100

    ax.text(
        p.get_x() + p.get_width()/2,
        count + total*0.005,
        f"{count:,} ({pct:.2f}%)",
        ha='center',
        fontsize=10
    )

plt.tight_layout()
plt.show()
```

**→ Nhận xét:** Trong nhóm tuổi trung niên (26–45), nam giới chiếm gần 68% tổng số va chạm nhẹ, gấp hơn hai lần so với nữ giới (32%). Điều này cho thấy nam tài xế trung niên – nhóm lao động chính, thường tham gia giao thông nhiều vào giờ cao điểm – là đối tượng trung tâm của các va chạm nhẹ.

---

Khi đã xác định rõ chân dung nhân khẩu học của nhóm tài xế chiếm ưu thế, câu hỏi tiếp theo là: ***Những nam tài xế trung niên này đang điều khiển loại phương tiện nào, và loại xe nào liên quan nhiều nhất đến các va chạm nhẹ?***

```python
# Lọc va chạm nhẹ, nhóm tuổi trung niên và nam giới
mid_age_male_data = merged_data[
    (merged_data['Accident_Severity'] == 'Slight') &
    (merged_data['Age_Band_of_Driver'].isin(['26 - 35', '36 - 45'])) &
    (merged_data['Sex_of_Driver'] == 'Male')
]

plt.figure(figsize=(10,5))

# Chọn top 5 loại phương tiện
top_vehicle = mid_age_male_data['Vehicle_Type'].value_counts().nlargest(5).index
ax = sns.countplot(
    data=mid_age_male_data[mid_age_male_data['Vehicle_Type'].isin(top_vehicle)],
    y='Vehicle_Type',
    order=top_vehicle[::1],
    color='#A0522D'
)

plt.title("Top 5 loại phương tiện của nam tài xế trung niên liên quan va chạm nhẹ", fontsize=14)
plt.xlabel("Số vụ va chạm nhẹ")
plt.ylabel("Loại phương tiện")

# Ghi số lượng + %
total = len(mid_age_male_data)
max_count = max(p.get_width() for p in ax.patches)

for p in ax.patches:
    count = int(p.get_width())
    pct = count / total * 100

    # Highlight cột dài nhất
    if count == max_count:
        p.set_color('#5C4033')  # nâu đậm

    ax.text(
        count + total*0.01,
        p.get_y() + p.get_height()/2,
        f"{count:,} ({pct:.2f}%)",
        va='center',
        fontsize=10
    )

plt.tight_layout()
plt.show()
```

**→ Nhận xét:** Dữ liệu về loại phương tiện cho thấy một kết quả rất rõ ràng **XE HƠI (Car)** là phương tiện chiếm ưu thế tuyệt đối với gần 136 ngàn vụ tương đương 69%, trong khi các loại xe khác như xe tải nhỏ xe buýt hay xe máy chỉ chiếm tỷ lệ rất nhỏ điều này khẳng định xe hơi không chỉ là phương tiện phổ biến nhất mà còn là loại xe liên quan chính đến các va chạm nhẹ.

---

Như vậy, qua phân tích đơn biến tập trung vào nhóm va chạm nhẹ **(Slight)**, có thể nhận thấy rằng các sự cố này chủ yếu xảy ra vào chiều tối các ngày làm việc, tập trung trên các tuyến đường đơn trong khu vực đô thị. Nhóm tài xế liên quan nhiều nhất là nam giới trong độ tuổi từ 26 đến 45, với phương tiện phổ biến là xe hơi. Điều này cho thấy mật độ giao thông cao, áp lực thời gian trong giờ cao điểm và hành vi lái xe trong môi trường đô thị là những yếu tố góp phần làm gia tăng số lượng các va chạm nhẹ. Tuy nhiên, câu hỏi đặt ra là: ***khi các yếu tố này cùng xuất hiện, yếu tố nào thực sự làm gia tăng rủi ro? Để trả lời câu hỏi đó, chúng ta chuyển sang phần phân tích đa biến.***

###**5.2.Phân tích đa biến (Multivariate Analysis)**

Sau khi xác định phần lớn tai nạn là va chạm nhẹ **(85%)**, chúng ta cần hiểu sâu hơn về cơ chế của chúng. Đồng thời, những vụ nghiêm trọng dù hiếm **(15%)** nhưng hậu quả nặng nề, cần được nghiên cứu riêng. Do đó, phân tích đa biến sẽ được chia thành hai dòng chảy song song:
+ NHÁNH A: TÌM HIỂU CƠ CHẾ CỦA CÁC VA CHẠM NHẸ **(Sligth).**
  + **A[1] - Kiểm tra tương tác giữa KHÔNG GIAN và THỜI GIAN:** Trong những ngày làm việc, liệu sự gia tăng va chạm vào giờ chiều tối có đặc biệt nghiêm trọng và tập trung ở khu vực đô thị so với nông thôn?
  + **A[2] - Kiểm tra tương tác giữa KHÔNG GIAN và CON NGƯỜI/PHƯƠNG TIỆN:** Trên các tuyến đường đơn trong đô thị – có đúng tổ hợp "nam tài xế 26-45 tuổi đi xe hơi" là nhóm chiếm ưu thế so với các tổ hợp khác cùng giới tính không?
+ NHÁNH B: TÌM HIỂU NGUYÊN NHÂN CỦA CÁC VỤ NGHIÊM TRỌNG **(Serious/Fatal).**
  + **B[1] - Thời gian chết chóc:** Liệu ban đêm có làm mỗi vụ tai nạn trở nên nguy hiểm hơn, với tỷ lệ tử vong cao hơn so với ban ngày hay không?
  + **B[2] - Thời tiết chết chóc:** Điều kiện thời tiết xấu (mưa, sương mù) có làm tăng đáng kể tỷ lệ tai nạn nghiêm trọng so với thời tiết tốt?
  + **B[3] - Con đường chết chóc:** Đường tốc độ cao (>60 km/h) có biến mỗi vụ tai nạn thành tai nạn nghiêm trọng hơn so với các đoạn đường tốc độ thấp hay không?
  + **B[4] - Con người chết chóc:** Khi một tai nạn xảy ra, độ tuổi tài xế có ảnh hưởng đến khả năng tai nạn đó trở thành nghiêm trọng hoặc tử vong hay không?
---

#### **(NHÁNH A) - TÌM HIỂU CƠ CHẾ CỦA CÁC VA CHẠM NHẸ (Slight)**

**A[1] - Kiểm tra tương tác giữa KHÔNG GIAN và THỜI GIAN:** Trong những ngày làm việc, liệu sự gia tăng va chạm vào giờ chiều tối có đặc biệt nghiêm trọng và tập trung ở khu vực đô thị so với nông thôn?

```python
# Lọc dữ liệu: va chạm nhẹ + ngày làm việc + khung giờ chiều tối
slight_evening = merged_data[
    (merged_data['Accident_Severity'] == 'Slight') &
    (merged_data['Day_of_Week'].isin(['Monday','Tuesday','Wednesday','Thursday','Friday'])) &
    (merged_data['Time_period'] == 'Chiều tối')
]

evening_counts = slight_evening['Urban_or_Rural_Area'].value_counts().sort_index()
slight_weekday_total = merged_data[
    (merged_data['Accident_Severity'] == 'Slight') &
    (merged_data['Day_of_Week'].isin(['Monday','Tuesday','Wednesday','Thursday','Friday']))
]
total_counts = slight_weekday_total['Urban_or_Rural_Area'].value_counts().sort_index()
evening_ratio = (evening_counts / total_counts * 100).round(1)

# Vẽ BIỂU ĐỒ
plt.figure(figsize=(8, 5))
bars = plt.bar(evening_counts.index, evening_counts.values, color=['#D6C1A3', '#6B4F3F'], width=0.6)

plt.title('Sự tập trung va chạm nhẹ vào CHIỀU TỐI theo khu vực\n(Thứ 2 - Thứ 6)',
          fontweight='bold', pad=15, fontsize=14)
plt.xlabel('Khu vực', fontsize=12)
plt.ylabel('Số vụ va chạm nhẹ vào chiều tối (T2-T6)', fontsize=12)

# Ghi số vụ + Tỷ lệ
for bar, count, ratio, total in zip(bars, evening_counts.values, evening_ratio.values, total_counts.values):
    height = bar.get_height()
    text = f'{count:,} vụ\n({ratio}% của {total:,})'
    plt.text(bar.get_x() + bar.get_width()/2., height + max(evening_counts.values)*0.02,
             text, ha='center', va='bottom', fontsize=10, fontweight='bold')

plt.grid(axis='y', alpha=0.3)
plt.ylim(0, max(evening_counts.values) * 1.25)
plt.tight_layout()
plt.show()
```

**→ Nhận xét:** Biểu đồ cho thấy một sự khác biệt đáng chú ý: tỷ lệ va chạm nhẹ xảy ra vào chiều tối tại đô thị **(34.5%)** cao hơn tại nông thôn **(32.1%)**. Mặc dù chênh lệch chỉ **2.4%**, nhưng điều này vẫn cho thấy nhịp điệu "chiều tối" có tác động mạnh hơn một chút trong môi trường đô thị. Kết hợp với số lượng tuyệt đối rất lớn (172,273 vụ), điều này củng cố rằng đô thị vào chiều tối thực sự là "điểm nóng" về mặt tần suất sự cố.

---

#### **(NHÁNH B) - TÌM HIỂU NGUYÊN NHÂN CỦA CÁC VỤ NGHIÊM TRỌNG (Serious/Fatal)**

Nhiều nghiên cứu cho thấy tai nạn vào ban đêm thường nghiêm trọng hơn do tầm nhìn hạn chế, tốc độ cao và khả năng phản ứng chậm. Trong bộ dữ liệu này, **B[1] - Thời gian chết chóc:** Liệu ban đêm có làm mỗi vụ tai nạn trở nên nguy hiểm hơn, với tỷ lệ tử vong cao hơn so với ban ngày hay không?

```python
df = merged_data[['Accident_Severity', 'Light_Conditions']].dropna()

severity_map = {'Slight': 'Slight', 'Serious': 'Serious', 'Fatal': 'Fatal'}
df['Severity_Label'] = df['Accident_Severity'].map(severity_map)

df_b = df[df['Severity_Label'].isin(['Serious', 'Fatal'])].copy()

def aggregate_light(condition):
    condition_str = str(condition)
    if condition_str == 'Daylight':
        return 'Ban ngày'
    else:
        return 'Ban đêm'

df_b['Light_Group'] = df_b['Light_Conditions'].apply(aggregate_light)
count_tab = pd.crosstab(df_b['Light_Group'], df_b['Severity_Label'])

severity_order = ['Serious', 'Fatal']
count_tab = count_tab[severity_order]

cross_tab = count_tab.div(count_tab.sum(axis=1), axis=0) * 100

group_order = ['Ban ngày', 'Ban đêm']
cross_tab = cross_tab.reindex(group_order)
count_tab = count_tab.reindex(group_order)

# VẼ BIỂU ĐỒ
plt.figure(figsize=(10, 7))

colors = ['#aab7b8', '#154360']
ax = cross_tab.plot(kind='bar', stacked=True, color=colors,
                    edgecolor='white', width=0.6, ax=plt.gca(), alpha=0.9)

for i in range(len(cross_tab)):
    cumulative_height = 0
    for j, severity in enumerate(severity_order):
        value = cross_tab.iloc[i, j]
        if value > 0:
            y_pos = cumulative_height + value / 2
            plt.text(i, y_pos, f'{value:.1f}%',
                     ha='center', va='center', color='white',
                     fontweight='bold', fontsize=11)
            cumulative_height += value

    n_serious = count_tab.iloc[i]['Serious']
    n_fatal = count_tab.iloc[i]['Fatal']
    total_n = n_serious + n_fatal
    header_text = f'n = {int(total_n):,} vụ\n({int(n_serious):,} Serious, {int(n_fatal):,} Fatal)'
    plt.text(i, 102, header_text,
             ha='center',
             va='bottom',
             fontweight='bold',
             color='#2d3436',
             fontsize=10,
             linespacing=1.5)

# Tinh chỉnh lại các thông số còn lại
plt.title('TỶ LỆ MỨC ĐỘ TAI NẠN NGHIÊM TRỌNG/TỬ VONG\n(Ban ngày vs Ban đêm)',
          fontsize=14, fontweight='bold')
plt.ylabel('Tỷ lệ phần trăm (%)', fontsize=12, fontweight='bold')
plt.xlabel('', fontsize=12)
plt.xticks(rotation=0, fontsize=12, fontweight='bold')
plt.grid(axis='y', alpha=0.3, linestyle='--')
plt.ylim(0, 110)

# Chú thích
plt.legend(title='Mức độ', loc='upper right', fontsize=8, frameon=True, shadow=True)

plt.tight_layout()
plt.show()
```

**→ Nhận xét:** Dựa trên biểu đồ, có thể khẳng định giả thuyết ban đầu là đúng: tai nạn xảy ra vào ban đêm nguy hiểm hơn rõ rệt so với ban ngày. **BAN NGÀY** có nhiều vụ nghiêm trọng hơn, nhưng **BAN ĐÊM** lại có tỷ lệ chết người cao hơn nhiều. Cụ thể, trong các vụ nghiêm trọng và tử vong, **BAN ĐÊM** có gần 10% là chết người, còn **BAN NGÀY** chỉ hơn 6%. Điều này có nghĩa cùng một vụ tai nạn, nếu xảy ra **BAN ĐÊM** thì khả năng dẫn đến tử vong cao hơn đáng kể so với **BAN NGÀY** - phản ánh mức độ rủi ro lớn hơn do điều kiện ánh sáng, tầm nhìn khi lái xe vào ban đêm.

Sau khi xác nhận ban đêm làm gia tăng mức độ nguy hiểm của mỗi vụ tai nạn, câu hỏi tiếp theo là liệu điều kiện môi trường bất lợi có tiếp tục khuếch đại rủi ro này hay không. **B[2] - Thời tiết chết chóc:** Điều kiện thời tiết xấu (mưa, sương mù) có làm tăng đáng kể tỷ lệ tai nạn nghiêm trọng so với thời tiết tốt?

```python
df = merged_data[['Accident_Severity', 'Weather_Conditions']].dropna().copy()

# Mapping
severity_map = {1: 'Fatal', 2: 'Serious', 3: 'Slight',
                'Fatal': 'Fatal', 'Serious': 'Serious', 'Slight': 'Slight'}
df['Severity_Label'] = df['Accident_Severity'].map(severity_map)
df_b2 = df[df['Severity_Label'].isin(['Serious', 'Fatal'])].copy()

def classify_weather(w):
    w = str(w).lower()
    if 'fine' in w:
        return 'Thời tiết tốt'
    else:
        return 'Thời tiết xấu'

df_b2['Thời tiết'] = df_b2['Weather_Conditions'].apply(classify_weather)

count_tab = pd.crosstab(df_b2['Thời tiết'], df_b2['Severity_Label'])
severity_order = ['Serious', 'Fatal']
count_tab = count_tab[severity_order]

cross_tab = count_tab.div(count_tab.sum(axis=1), axis=0) * 100
group_order = ['Thời tiết tốt', 'Thời tiết xấu']
cross_tab = cross_tab.reindex(group_order)
count_tab = count_tab.reindex(group_order)

plt.figure(figsize=(10, 7))
colors = ['#FFAB91', '#B71C1C']

ax = cross_tab.plot(kind='bar', stacked=True, color=colors,
                    edgecolor='white', width=0.6, ax=plt.gca(), alpha=0.9)

# Chèn tỷ lệ % vào lòng cột và số liệu chi tiết lên đỉnh
for i in range(len(cross_tab)):
    cum_height = 0
    for j, sev in enumerate(severity_order):
        val = cross_tab.iloc[i, j]
        if val > 0:
            plt.text(i, cum_height + val/2, f'{val:.1f}%',
                     ha='center', va='center', color='white', fontweight='bold', fontsize=11)
            cum_height += val

    n_s = count_tab.iloc[i]['Serious']
    n_f = count_tab.iloc[i]['Fatal']
    plt.text(i, 102, f'n = {int(n_s+n_f):,} vụ\n({int(n_s):,} S, {int(n_f):,} F)',
             ha='center', va='bottom', fontweight='bold', color='#2d3436', fontsize=9)

plt.title('TỶ LỆ MỨC ĐỘ TAI NẠN NGHIÊM TRỌNG/TỬ VONG\n(Thời tiết tốt vs Thời tiết xấu)',
          fontsize=13, fontweight='bold')
plt.ylabel('Tỷ lệ phần trăm (%)', fontweight='bold')
plt.xticks(rotation=0, fontweight='bold')
plt.ylim(0, 110)
plt.grid(axis='y', alpha=0.2, linestyle='--')

# Chú thích góc trên phải
plt.legend(title='Mức độ', loc='upper right', frameon=True, shadow=True)

plt.tight_layout()
plt.show()
```

**→ Nhận xét:** Dữ liệu này phủ nhận giả thuyết về "thời tiết chết chóc". Tỷ lệ tử vong trong điều kiện **THỜI TIẾT XẤU** chỉ là 7.3%, gần như không khác biệt so với 7.5% trong **THỜI TIẾT TỐT**. Khi so sánh với phát hiện trước đó về ánh sáng - nơi ban đêm làm tăng nguy cơ tử vong lên 51% - thì rõ ràng bóng tối mới là yếu tố nguy hiểm thực sự, còn mưa hay sương mù không làm thay đổi đáng kể khả năng một vụ tai nạn sẽ gây chết người.

---

Sau khi thấy thời tiết không phải là “kẻ giết người” như giả thuyết đặt ra, câu hỏi tiếp theo là liệu tốc độ cho phép của đoạn đường có phải là yếu tố quyết định mức độ nghiêm trọng của hậu quả tai nạn hay không. **B[3] - Con đường chết chóc:** Đường tốc độ cao (>60 km/h) có biến mỗi vụ tai nạn thành tai nạn nghiêm trọng hơn so với các đoạn đường tốc độ thấp hay không?

```python
df = merged_data[['Accident_Severity', 'Speed_limit']].dropna().copy()

severity_map = {1: 'Fatal', 2: 'Serious', 3: 'Slight',
                'Fatal': 'Fatal', 'Serious': 'Serious', 'Slight': 'Slight'}
df['Severity_Label'] = df['Accident_Severity'].map(severity_map)
df_b3 = df[df['Severity_Label'].isin(['Serious', 'Fatal'])].copy()

def classify_speed(speed):
    try:
        s = float(speed)
        return 'Đường tốc độ cao (>60 km/h)' if s > 60 else 'Đường tốc độ thấp (≤60 km/h)'
    except:
        return 'Không rõ'

df_b3['Tốc độ'] = df_b3['Speed_limit'].apply(classify_speed)

count_tab = pd.crosstab(df_b3['Tốc độ'], df_b3['Severity_Label'])
severity_order = ['Serious', 'Fatal']
count_tab = count_tab[severity_order]

cross_tab = count_tab.div(count_tab.sum(axis=1), axis=0) * 100
group_order = ['Đường tốc độ thấp (≤60 km/h)', 'Đường tốc độ cao (>60 km/h)']
cross_tab = cross_tab.reindex(group_order)
count_tab = count_tab.reindex(group_order)

# VẼ BIỂU ĐỒ
plt.figure(figsize=(10, 7))
colors = ['#F1948A', '#4A235A'] # Serious (Hồng cam), Fatal (Đỏ đậm)

ax = cross_tab.plot(kind='bar', stacked=True, color=colors,
                    edgecolor='white', width=0.6, ax=plt.gca(), alpha=0.9)

# Chèn tỷ lệ % vào lòng cột và số liệu chi tiết lên đỉnh
for i in range(len(cross_tab)):
    cum_height = 0
    for j, sev in enumerate(severity_order):
        val = cross_tab.iloc[i, j]
        if val > 0:
            plt.text(i, cum_height + val/2, f'{val:.1f}%',
                     ha='center', va='center', color='white', fontweight='bold', fontsize=11)
            cum_height += val

    n_s = count_tab.iloc[i]['Serious']
    n_f = count_tab.iloc[i]['Fatal']
    plt.text(i, 102, f'n = {int(n_s+n_f):,} vụ\n({int(n_s):,} S, {int(n_f):,} F)',
             ha='center', va='bottom', fontweight='bold', color='#2d3436', fontsize=9)

plt.title('TỶ LỆ MỨC ĐỘ TAI NẠN NGHIÊM TRỌNG/TỬ VONG\n(Tốc độ thấp vs Tốc độ cao)',
          fontsize=13, fontweight='bold')
plt.ylabel('Tỷ lệ phần trăm (%)', fontweight='bold')
plt.xticks(rotation=0, fontweight='bold')

plt.ylim(0, 110)
plt.grid(axis='y', alpha=0.2, linestyle='--')
plt.legend(title='Mức độ', loc='upper right', bbox_to_anchor=(1, 1),
           frameon=True, shadow=True, borderaxespad=0)

plt.tight_layout()
plt.show()
```

**→ Nhận xét:** Dựa trên biểu đồ, có thể khẳng định giả thuyết ban đầu là đúng: tai nạn xảy ra trên đường tốc độ cao nguy hiểm hơn rõ rệt so với đường tốc độ thấp. **ĐƯỜNG TỐC ĐỘ THẤP** có nhiều vụ nghiêm trọng hơn, nhưng **ĐƯỜNG TỐC ĐỘ CAO** lại có tỷ lệ chết người cao hơn gấp đôi. Cụ thể, trong các vụ nghiêm trọng và tử vong, **ĐƯỜNG TỐC ĐỘ CAO** có đến 14.2% là chết người, còn **ĐƯỜNG TỐC ĐỘ THẤP** chỉ 7.0%. Điều này có nghĩa cùng một vụ tai nạn, nếu xảy ra trên **ĐƯỜNG TỐC ĐỘ CAO** thì khả năng dẫn đến tử vong cao hơn gấp 2 lần so với **ĐƯỜNG TỐC ĐỘ THẤP** - phản ánh mức độ rủi ro cực lớn do động năng cao và thời gian phản ứng hạn chế khi di chuyển ở tốc độ lớn.

---

Khi đã đi qua các yếu tố môi trường và hạ tầng, vậy còn độ tuổi tài xế có là yếu tố ảnh hưởng đáng kể đến mức độ nghiêm trọng của tai nạn hay không? **B[4] - Con người chết chóc:** Khi một tai nạn xảy ra, độ tuổi tài xế có ảnh hưởng đến khả năng tai nạn đó trở thành nghiêm trọng hoặc tử vong hay không?

```python
df = merged_data[['Accident_Severity', 'Age_Band_of_Driver']].dropna().copy()
df = df[df['Age_Band_of_Driver'] != 'Missing']

severity_map = {
    1: 'Fatal', 2: 'Serious', 3: 'Slight',
    'Fatal': 'Fatal', 'Serious': 'Serious', 'Slight': 'Slight'
}
df['Severity_Label'] = df['Accident_Severity'].map(severity_map)

df_b4 = df[df['Severity_Label'].isin(['Serious', 'Fatal'])].copy()

# GOM 3 NHÓM ĐỘ TUỔI
def classify_age_3groups(age):
    age = str(age)
    # Nhóm trẻ ≤25
    if any(x in age for x in ['0 - 5', '6 - 10', '11 - 15', '16 - 20', '21 - 25']):
        return 'Tài xế trẻ (≤25)'
    # Nhóm trung niên 26–65
    elif any(x in age for x in ['26 - 35', '36 - 45', '46 - 55', '56 - 65']):
        return 'Tài xế trung niên (26–65)'
    # Nhóm lớn tuổi ≥66
    elif any(x in age for x in ['66 - 75', 'Over 75']):
        return 'Tài xế lớn tuổi (≥66)'
    else:
        return None

df_b4['Nhóm tuổi'] = df_b4['Age_Band_of_Driver'].apply(classify_age_3groups)
df_b4 = df_b4.dropna(subset=['Nhóm tuổi'])

count_tab = pd.crosstab(df_b4['Nhóm tuổi'], df_b4['Severity_Label'])
severity_order = ['Serious', 'Fatal']
count_tab = count_tab[severity_order]
cross_tab = count_tab.div(count_tab.sum(axis=1), axis=0) * 100

group_order = [
    'Tài xế trẻ (≤25)',
    'Tài xế trung niên (26–65)',
    'Tài xế lớn tuổi (≥66)'
]
cross_tab = cross_tab.reindex(group_order)
count_tab = count_tab.reindex(group_order)

# VẼ BIỂU ĐỒ
plt.figure(figsize=(12, 7))

colors_b4 = ['#80CBC4', '#BF360C']

ax = cross_tab.plot(
    kind='bar', stacked=True, color=colors_b4,
    edgecolor='white', width=0.8, ax=plt.gca(), alpha=0.9
)

# Ghi % trong cột
for i in range(len(cross_tab)):
    cum_height = 0
    for j, sev in enumerate(severity_order):
        val = cross_tab.iloc[i, j]
        if val > 0:
            plt.text(
                i, cum_height + val / 2,
                f'{val:.1f}%',
                ha='center', va='center',
                color='white', fontweight='bold', fontsize=11
            )
            cum_height += val

    # Ghi số lượng
    n_s = count_tab.iloc[i]['Serious']
    n_f = count_tab.iloc[i]['Fatal']
    plt.text(
        i, 102,
        f'n = {int(n_s + n_f):,} vụ\n({int(n_s):,} S, {int(n_f):,} F)',
        ha='center', va='bottom',
        fontweight='bold', fontsize=9, color='#2d3436'
    )

plt.title('TỶ LỆ TAI NẠN NGHIÊM TRỌNG & TỬ VONG THEO NHÓM TUỔI TÀI XẾ\n',
          fontsize=13, fontweight='bold')
plt.ylabel('Tỷ lệ (%)', fontweight='bold')
plt.xticks(rotation=0, fontweight='bold')
plt.ylim(0, 110)
plt.grid(axis='y', linestyle='--', alpha=0.2)

plt.legend(title='Mức độ',loc='upper right',frameon=True, shadow=True)
plt.tight_layout()
plt.show()
```

**→ Nhận xét:** Dựa trên biểu đồ, có thể thấy rõ độ tuổi tài xế có ảnh hưởng đến khả năng tai nạn trở thành tử vong. Tài xế lớn tuổi (≥66) có tỷ lệ tử vong cao nhất (9.8%), tiếp theo là trung niên (8.0%) và thấp nhất là nhóm trẻ (7.6%). Điều này cho thấy cùng một vụ tai nạn, nếu tài xế là người lớn tuổi thì khả năng dẫn đến tử vong cao hơn đáng kể so với các nhóm tuổi khác, phản ánh mức độ dễ bị tổn thương về thể chất cao hơn ở người cao tuổi khi xảy ra va chạm.

##**6. Kết luận**

Bức tranh tai nạn giao thông cho thấy mức độ nghiêm trọng không phải là ngẫu
nhiên, mà được hình thành bởi hai cơ chế rủi ro khác nhau nhưng bổ trợ lẫn nhau, đó là cơ chế tạo ra nhiều va chạm nhẹ và cơ chế quyết định tai nạn trở nên nghiêm trọng hoặc tử vong.

**Cơ chế tạo ra va chạm nhẹ – câu chuyện của tần suất và mật độ giao thông:**
Các va chạm nhẹ chiếm số lượng áp đảo trong dữ liệu và chủ yếu gắn liền với
cường độ tham gia giao thông cao hơn là mức độ nguy hiểm của bối cảnh. Tai nạn
dạng này tập trung vào các ngày làm việc, khung giờ chiều tối, đặc biệt trên các
tuyến đường đơn trong khu vực đô thị. Nhóm tài xế nam trong độ tuổi lao động,
sử dụng ô tô cá nhân, xuất hiện với tần suất cao, phản ánh vai trò của nhóm người di chuyển thường xuyên trong môi trường giao thông có mật độ lớn.
Nguyên nhân cốt lõi của cơ chế này là áp lực giao thông đô thị, mật độ phương
tiện lớn và các tương tác phức tạp giữa các dòng xe trong giờ cao điểm, dẫn đến
nhiều va chạm nhỏ nhưng khó tránh khỏi.

**→ Giải pháp** phù hợp cho nhóm tai nạn này cần tập trung vào quản lý giao thông đô thị, bao gồm tổ chức luồng xe hợp lý, tối ưu hóa hệ thống tín hiệu, điều tiết giờ cao điểm và giảm xung đột tại các nút giao. Những biện pháp này giúp giảm số lượng va chạm, dù không trực tiếp tác động mạnh đến tỷ lệ tử vong.

**Cơ chế quyết định tai nạn nghiêm trọng và tử vong – câu chuyện của bối
cảnh và khả năng chịu va chạm:**
Trái ngược với cơ chế thứ nhất, việc một vụ tai nạn trở nên nghiêm trọng hay
dẫn đến tử vong lại chịu chi phối mạnh bởi bối cảnh xảy ra tai nạn và khả năng
chịu va chạm của con người. Tai nạn xảy ra trên các tuyến đường có tốc độ giới
hạn cao, khu vực thưa dân, vào ban đêm và trong điều kiện ánh sáng kém có xác
suất gây hậu quả nghiêm trọng cao hơn rõ rệt. Trong cùng một loại va chạm, tai
nạn xảy ra trong các bối cảnh này thường gây tổn thương nặng nề hơn so với tai
nạn ban ngày trong đô thị. Nguyên nhân cốt lõi của cơ chế này là năng lượng va chạm lớn, thời gian phản
ứng kém, khả năng quan sát hạn chế và mức độ dễ tổn thương cao hơn của người
tham gia giao thông trong các bối cảnh bất lợi.
